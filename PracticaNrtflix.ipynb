### Me doy cuenta que el dataset tiene algunos datos faltantes, los eliminare

```python
#Contar los valores faltantes
df.isnull().sum()
#mostrar el porcentaje de valores faltantes
df.isnull().sum().sum()/len(df)*100
```

```python
## eliminar los datos nulos

#eliminar los valores faltantes
df.dropna(inplace=True)
```

```python
## veo de nuevo como quedo mi dataset
df.info()
```

### Me gustaria asegurarme que mi dataset no tenga filas repetidas


```python
df = df.drop_duplicates()
df
```

### Eliminar Registros con Datos Incorrectos o Poco Útiles

```python
df = df[(df["release_year"] >= 1900) & (df["release_year"] <= 2025)]
```

### Detectar y Corregir Países con Nombres Duplicados

```python
country_mapping = {
    "United States": "USA",
    "United Kingdom": "UK",
    "South Korea": "Korea",
}

df["country"] = df["country"].replace(country_mapping)
```

### Verificar Valores Únicos y Errores en Categorías

```python
print(df["rating"].unique())  # Muestra todas las clasificaciones por edad
rating_mapping = {
    "UR": "Unrated",
    "NC-17": "Adults Only",
    "TV-Y7-FV": "TV-Y7"
}

df["rating"] = df["rating"].replace(rating_mapping)
df
```

### Visualizar Outliers con Boxplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["duration"])
plt.title("Boxplot de Duración (Minutos/Temporadas)")
plt.show()
```
